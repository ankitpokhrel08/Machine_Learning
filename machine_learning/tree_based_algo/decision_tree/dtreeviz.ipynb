```python
import graphviz.backend as be
from sklearn.datasets import *
from dtreeviz.trees import *
from IPython.display import Image, display_svg, SVG
```

```python
X,y = load_iris(return_X_y=True)
dtc = DecisionTreeClassifier(max_depth=3)
dtc.fit(X, y)
```

```python
#Visualizing using plot_tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
tree.plot_tree(dtc, filled=True)
plt.show()
```

```python
from dtreeviz import model

# Using the modern dtreeviz API (version 2.0+)
viz_model = model(
    dtc,
    X_train=X,
    y_train=y,
    feature_names=load_iris().feature_names,
    class_names=load_iris().target_names
)
viz = viz_model.view()
viz
```

```python
# Create random sample and visualize using modern dtreeviz API with path tracing
sample_X = iris.data[np.random.randint(0, len(iris.data)), :]

# Predict what class this sample should be
predicted_class = dtc.predict([sample_X])[0]
class_names = ["setosa", "versicolor", "virginica"]
print(f"Predicted class: {class_names[predicted_class]}")

# Use iris.data and iris.target (the full dataset)
from dtreeviz import model

# Create the model using iris.data and iris.target (full dataset)
viz_model = model(
    dtc,  # Using dtc classifier from cell 4
    X_train=iris.data,  # Using full iris dataset
    y_train=iris.target,  # Using full iris target
    feature_names=iris.feature_names,
    class_names=["setosa", "versicolor", "virginica"]
)

# Create visualization WITH PATH TRACING for the selected sample
viz = viz_model.view(x=sample_X)  # This will highlight the path through the tree
viz
```
