```python
#Creating nonlinear toy dataset using make_moons
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)
```

```python
#plotting the dataset
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Nonlinear Toy Dataset - make_moons')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Trying to fit logistic regression on it without polynomial features
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)
```

```python
#Plotting decision boundary
import numpy as np
xx, yy = np.meshgrid(np.linspace(-1.5, 2.5, 100), np.linspace(-1, 1.5, 100))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Logistic Regression without Polynomial Features')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Using Polynomial Features to transform the data
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=5)
X_poly = poly.fit_transform(X)  
```

```python
#Fitting logistic regression on transformed data
model_poly = LogisticRegression()
model_poly.fit(X_poly, y)
```

```python
#Plotting decision boundary
xx, yy = np.meshgrid(np.linspace(-1.5, 2.5, 100), np.linspace(-1, 1.5, 100))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
Z_poly = model_poly.predict(poly.transform(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)
plt.contourf(xx, yy, Z_poly, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Logistic Regression with Polynomial Features')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
