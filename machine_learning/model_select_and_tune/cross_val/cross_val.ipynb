```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
```

```python
#Using KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
### Using Cross-Validation to evaluate the model
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
scores = cross_val_score(knn, X, y, cv=loo)
print(f"Mean Cross-Validation Score: {np.mean(scores)}")
```

```python
#Importing dibetics dataset from sklearn
from sklearn.datasets import load_diabetes
X,y = load_diabetes(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
#Calculating Mean Squared Error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
#Using Cross Validation to evaluate a model
### Using Cross-Validation to evaluate the model
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
scores = cross_val_score(lr, X, y, cv=loo, scoring='neg_mean_squared_error')
print(f"Mean Cross-Validation MSE: {np.mean(-scores)}")
```

```python
### K-Folds Cross-Validation
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=44)
kf_scores = cross_val_score(lr, X, y, cv=kf, scoring='neg_mean_squared_error')
#R2 scoring can also be used for regression tasks
r2_scores = cross_val_score(lr, X, y, cv=kf, scoring='r2')
print(f"K-Folds Cross-Validation R2 Scores: {r2_scores}")
print(f"Mean K-Folds Cross-Validation R2 Score: {np.mean(r2_scores)}")

print(f"K-Folds Cross-Validation Scores: {kf_scores}")
print(f"Mean K-Folds Cross-Validation Score: {np.mean(-kf_scores)}")
```

```python
#Using KNN Classifier
from sklearn.datasets import load_breast_cancer
X,y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
#Accuracy Score
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# Using Stratified K-Folds for Classification
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=44)
skf_scores = cross_val_score(knn, X, y, cv=skf,scoring='accuracy')
print(f"Stratified K-Folds Cross-Validation Scores: {skf_scores}")
print(f"Mean Stratified K-Folds Cross-Validation Score: {np.mean(skf_scores)}")
```
