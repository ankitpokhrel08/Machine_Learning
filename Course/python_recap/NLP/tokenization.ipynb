```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Nepali sentence
sentence = "“नेपाल एक अद्भुत देश हो जुन प्राकृतिक सुन्दरतामा भरिएको छ। यहाँका हिमाल, नदियाँ, र हरियाली खेतहरू देख्न लाज नलाग्ने खालका छन्। नेपालमा विभिन्न जाति, भाषा र संस्कृतिका मानिसहरू बसोबास गर्छन्। तिनीहरूको आपसी सम्बन्ध र साझा जीवनशैली हाम्रो देशको सौहार्द र सद्भावलाई देखाउँछ। नेपालको समृद्धि र विकासका लागि हामी सबैले मिलेर काम गर्नुपर्छ।”"

# Tokenization
words = word_tokenize(sentence)

# Loading Nepali stopwords (if available)
stop_words = stopwords.words('nepali')

# Filtering stopwords
filtered_words = [word for word in words if word not in stop_words]

print("Original Sentence: ", sentence)
print("Tokenized Words: ", words)
print("Filtered Words (Stopwords Removed): ", filtered_words)
```
