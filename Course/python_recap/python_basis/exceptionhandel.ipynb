### Exception Handeling

### Common Python Exceptions

1. **Exception**: Base class for all exceptions.
2. **ArithmeticError**: Base class for all errors that occur for numeric calculation.
    - **OverflowError**: Raised when a calculation exceeds maximum limit for a numeric type.
    - **ZeroDivisionError**: Raised when division or modulo by zero takes place for all numeric types.
    - **FloatingPointError**: Raised when a floating point calculation fails.
3. **AssertionError**: Raised in case of failure of the Assert statement.
4. **AttributeError**: Raised in case of failure of attribute reference or assignment.
5. **EOFError**: Raised when there is no input from either the raw_input() or input() function and the end of file is reached.
6. **ImportError**: Raised when an import statement fails.
7. **KeyboardInterrupt**: Raised when the user interrupts program execution, usually by pressing Ctrl+C.
8. **IndexError**: Raised when an index is not found in a sequence.
9. **KeyError**: Raised when the specified key is not found in the dictionary.
10. **NameError**: Raised when an identifier is not found in the local or global namespace.
11. **MemoryError**: Raised when an operation runs out of memory.
12. **OSError**: Base class for all operating system-related errors.
     - **FileNotFoundError**: Raised when a file or directory is requested but doesnâ€™t exist.
     - **PermissionError**: Raised when trying to run an operation without the adequate access rights.
13. **TypeError**: Raised when an operation or function is attempted that is invalid for the specified data type.
14. **ValueError**: Raised when the built-in function for a data type has the valid type of arguments, but the arguments have invalid values specified.
15. **RuntimeError**: Raised when a generated error does not fall into any category.
16. **StopIteration**: Raised by the next() function to indicate that there is no further item to be returned by the iterator.
17. **SyntaxError**: Raised when there is an error in Python syntax.
18. **IndentationError**: Raised when there is an incorrect indentation.
     - **TabError**: Raised when the indentation consists of inconsistent tabs and spaces.
19. **SystemError**: Raised when the interpreter finds an internal problem, but when this error is encountered the Python interpreter does not exit.
20. **SystemExit**: Raised when Python interpreter is quit by using the sys.exit() function. If not handled in the code, causes the interpreter to exit.
21. **UnboundLocalError**: Raised when a reference is made to a local variable in a function or method, but no value has been bound to that variable.
22. **UnicodeError**: Raised when a Unicode-related encoding or decoding error occurs.
     - **UnicodeEncodeError**: Raised when a Unicode-related error occurs during encoding.
     - **UnicodeDecodeError**: Raised when a Unicode-related error occurs during decoding.
     - **UnicodeTranslateError**: Raised when a Unicode-related error occurs during translating.

```python
### try,except and else block
#Real life secenerio
try:
    num = int(input("Enter any number: "))
    result = 10/num
except ValueError as ve:
    print("Enter numbers only")
except ZeroDivisionError as ze:
    print("You cannot enter Zero")
except Exception as ex:
    print(ex)
else:
    print("Code ran without any exception and the result is",result)
```

```python
### try,except,else and finally
#Real life secenerio
try:
    num = int(input("Enter any number: "))
    result = 10/num
except ValueError as ve:
    print("Enter numbers only")
except ZeroDivisionError as ze:
    print("You cannot enter Zero")
except Exception as ex:
    print(ex)
else:
    print("Code ran without any exception and the result is",result)
finally:
    print("Execution completed!")
```
