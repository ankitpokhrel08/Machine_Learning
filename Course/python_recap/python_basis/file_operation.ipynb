```python
#Read a whole file content
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
#Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ##strip() removes the newline character
```

```python
import requests
response = requests.get("https://www.google.com")
with open("google.html",'w') as file:
    for line in response.text:
        file.write(line.strip())
```

```python
import requests

response = requests.get("https://www.google.com")

# Write HTML content line by line
with open("google.html", 'w', encoding='utf-8') as file:
    for line in response.text.splitlines():
        file.write(line + '\n')
```

```python
import requests

# Fetch the HTML content of the webpage
response = requests.get("https://x.com/")

# Save the HTML content to a file
with open("twitter.html", 'w', encoding='utf-8') as file:
    file.write(response.text)
```

```python
#writing line in files form list
lines = ['First line\n','Second line\n']
with open("newmethod.txt","w") as file:
    # for line in lines:
    #     file.write(line)
    file.writelines(lines)
```

```python
#Writing in binary file
data = b"\x00\x02"
with open("example.bin",'wb') as file:
    file.write(data)
#reading is with "rb"
```

```python
### Read the content from a source text file and write to a destination
with open("example.txt","r") as file:
    content = file.read()
with open("destination.md","w") as dest:
    dest.write(content)
```
