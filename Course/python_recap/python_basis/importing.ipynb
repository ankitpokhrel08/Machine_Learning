```python
## Data Serialization
import json
data = {'name':'Rakul','age':23}
#dumps will convert json into string
jsondata = json.dumps(data)
print(jsondata)
print(type(jsondata))
#loads will convert string into json
print(json.loads(jsondata))
print(type(json.loads(jsondata)))
#
```

```python
import csv 
with open('example.csv',mode='w',newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Rakul','45'])
    writer.writerow(['Rupesh','21'])
with open("example.csv",mode = 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
