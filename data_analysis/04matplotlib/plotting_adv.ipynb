```python
df =  pd.read_csv("datasets/batters.csv")
newdf = df.head(100).sample(25,random_state=6)
```

```python
newdf
```

```python
plt.figure(figsize=(15,7))
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.scatter(newdf['avg'],newdf['strike_rate'],s = newdf['runs'])
for i in range(newdf.shape[0]):
    plt.text(newdf['avg'].values[i],newdf['strike_rate'].values[i],newdf['batter'].values[i])
plt.show()
```

```python
#Subs-plots
fig,ax = plt.subplots()
ax.scatter(newdf['avg'],newdf['strike_rate'])
ax.set_title("Avg vs Strike Rate")
ax.set_ylabel("Strike Rate")
ax.set_xlabel("Avg")

fig.show()
```

```python
#Plotting conture surface plot
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
zz = xx**2 + yy**2 
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy,zz,color='aqua')
plt.savefig("figure.png")
```

```python
#Plotting conture surface plot
x = np.linspace(5,10,100)
y = np.linspace(5,10,100)
xx,yy = np.meshgrid(x,y)
zz = np.sin(xx**1.5)+np.cos(xx)
fig = plt.figure(figsize = (8,8))
ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy,zz,cmap='viridis')
ax.set_title("Graph of f(x,y) = sin(x^1.5)+ cos(x) ")
plt.savefig("plots.png")
```

```python
#Contour Plot
#Plotting conture surface plot
x = np.linspace(-100,10,100)
y = np.linspace(-100,10,100)
xx,yy = np.meshgrid(x,y)
```

```python
zz = xx**2 + yy**2
zz.shape
```

```python
fig = plt.figure()
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,zz, cmap = 'viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,zz,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
#HITMAP
df = pd.read_csv("datasets/ballbyball.csv")
df.head()
```

```python
df['ballnumber'].unique()
```

```python
temp_df = df[(df['batsman_run'] == 6) & (df['ballnumber']<=6)]
```

```python
grid = temp_df.pivot_table(index = 'overs', columns = 'ballnumber', values = 'batsman_run',aggfunc = 'count')
grid = grid.fillna(0)
grid
```

```python
# plt.figure(figsize = (10,10))
plt.imshow(grid)
plt.title("6's per Ball in IPL")
plt.xlabel("Ball")
plt.ylabel("Over")
plt.colorbar()
plt.yticks(df['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.savefig("sixes.png")
```
