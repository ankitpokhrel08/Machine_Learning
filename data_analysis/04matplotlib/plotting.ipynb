```python
# Pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes,explode = [0.1,0.1,0.1,0.1], labels=labels, autopct='%1.1f%%', startangle=90)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
#Plotting Entropy graph
a = np.linspace(0.1,0.9,20)
b = 1-a
ent = -a * np.log2(a) - b * np.log2(b)
```

```python
plt.plot(a,ent,color = "red")
plt.title("Entropy Graph")
plt.xlabel("Probability")
plt.ylabel("Entropy")
```
