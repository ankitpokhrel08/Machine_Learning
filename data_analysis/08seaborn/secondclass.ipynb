```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
iris.head()
```

```python
sns.stripplot(data=tips,x='day',y='tip',jitter=False)
```

```python
#Figure level graph
sns.catplot(data=tips,x='day',y='tip',kind='strip',hue='sex')
```

```python
#Boxplots
sns.boxplot(data=tips,x='day',y='tip')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.boxplot(data=tips,y='total_bill',hue='smoker')
```

```python
sns.violinplot(data=tips,y='total_bill',hue='day')
```

```python
sns.barplot(data=tips,y='total_bill',hue='day')
```

```python
sns.pointplot(data=tips,y='total_bill',hue='day')
```

```python
sns.pointplot(data=tips,x='sex',y='total_bill')
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',ci=None)
```

```python
sns.catplot(data=tips,x='sex',hue='day',col='smoker',kind='swarm',row='time')
```

```python
#RegPlot
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
#RegPlot
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
g =sns.lmplot(data=tips,x='total_bill',y='tip',hue='day')
g.fig.suptitle('Reg. Plot for Tips Dataset')
```

```python
g = sns.FacetGrid(data=tips,col='smoker',row='sex')
g.map(sns.scatterplot,'tip','total_bill')
```

```python
g = sns.pairplot(iris,hue='species')
g.fig.suptitle("Iris Dataset Visualisation")
```
