```python
gap = px.data.gapminder()
gap.head()
```

```python
sns.relplot(data=nepal,x='year',y='lifeExp',kind = 'line')
```

```python
sns.lineplot(data=nepal,x='year',y='lifeExp')
```

```python
nations = gap[gap['country'].isin(['India','Nepal','China'])]
```

```python
sns.lineplot(data=nations,x='year',y='lifeExp',hue='country')
```

```python
sns.relplot(data=nations,x='year',y='lifeExp',hue='country',kind='line')
```
