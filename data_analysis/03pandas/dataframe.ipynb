```python
points = [
    [(1,2),(2,3)],
    [(4,5),(7,8)],
    [(5,6),(-1,2)],
    [(9,3),(3,4)]
        ]
```

```python
#Calculate distance between two points
dist = pd.DataFrame( points, columns=['1_point','2_point'])
dist
```

```python
def distance(row):
    first = row['1_point']
    second = row['2_point']
    return ((second[1]-first[0])**2 + (second[0]-first[0])**2)**0.5
```

```python
dist.apply(distance,axis = 1)
```
