```python
#punctuations
import string
string.punctuation
```

```python
def remove_punctuation(data):
    for i in string.punctuation:
        if i in data:
            data = data.replace(i,"")
    return data
```

```python
df['review'] = df['review'].apply(remove_punctuation)
```

```python
df['tokenized_review'] = df['review'].apply(word_tokenize)
```

```python
# stop word removal
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    
    L = []
    for word in text:
        if word not in stopwords.words('english'):
            L.append(word)
            
    return L
```

```python
df['tokenized_review'] = df['tokenized_review'].apply(remove_stopwords)
```

```python
positive_words = set(opinion_lexicon.positive())  # convert list to set

def count_good(data):
    return sum(1 for word in data if word in positive_words)
```

```python
df['good'] = df['tokenized_review'].apply(count_good)
```

```python
negative_words = set(opinion_lexicon.negative())
def count_bad(data):
    return sum(1 for word in data if word in negative_words)
```

```python
df['bad'] = df['tokenized_review'].apply(count_bad)
```
