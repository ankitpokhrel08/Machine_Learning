```python
genres = movies.groupby("Genre")
```

```python
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
def find_rank(group):
    group['genre_rank']=group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
movies.groupby('Genre').apply(find_rank)
```

```python
#find normalized rating of each movie group wise
def normalised(group):
    group['Normalized_rating'] = (group['IMDB_Rating'] - (group['IMDB_Rating'].min()))/((group['IMDB_Rating'].max())-(group['IMDB_Rating'].min()))
    return group
    
```

```python
genres.apply(normalised)
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl.sample(4)
```

```python
ipl.shape
```

```python
#Top ten batsman with highest run
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman').size().sort_values(ascending=False).head(12)
```

```python
last5over = ipl[(ipl['over'] >= 15) & ((ipl['batsman_runs']== 6) | (ipl['batsman_runs']== 4 ))]
# last5over.groupby('batsman').size().sort_values(ascending=False)
last5over.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
#V Kohli stats against all teams
vk = ipl[ipl['batsman'] == 'V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
#Create a function that returns batsman highest score
def highscore(bat):
    return (ipl[ipl['batsman'] == bat]).groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1).values[0]

int(highscore('AB de Villiers'))
```
