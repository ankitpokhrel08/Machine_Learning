```python
import numpy as np
import pandas as pd
```

```python
matches = pd.read_csv('matches.csv')
matches = matches[matches['season'] == 2016]
newdata = pd.read_csv('deliveries.csv')
ballbyball = matches.merge(newdata,left_on = 'id',right_on = 'match_id')
vk = ballbyball[ballbyball['batsman'] == 'V Kohli'][['id','season','batsman','batsman_runs']]
vk = vk.groupby('id')['batsman_runs'].sum().reset_index()
vk = vk.merge(matches,on = 'id')[['venue','team1','team2','winner','player_of_match','batsman_runs','season']]
vk
```

```python
confirm = pd.read_csv('time_series_covid19_confirmed_global.csv')
death = pd.read_csv('time_series_covid19_deaths_global.csv')
```

```python
confirm
```

```python
death
```

```python
death = death.melt(id_vars = ['Province/State',	'Country/Region',	'Lat',	'Long'],var_name = "Date",value_name = "No_of_death")
confirm = confirm.melt(id_vars = ['Province/State',	'Country/Region',	'Lat',	'Long'],var_name = "Date",value_name = "No_of_confirm")
```

```python
covid = confirm.merge(death, on=['Province/State','Country/Region','Lat','Long','Date'])
```

```python
covid.sort_values('No_of_death' ,ascending = False)
```
