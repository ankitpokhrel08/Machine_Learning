```python
import re
import nepali_datetime

# Nepali month mapping
nepali_months = {
    "बैशाख": 1, "जेठ": 2, "असार": 3, "साउन": 4,
    "भदौ": 5, "असोज": 6, "कार्तिक": 7, "मंसिर": 8,
    "पुष": 9, "माघ": 10, "फागुन": 11, "चैत": 12
}

def nepali_date_to_ad(text):
    # Extract BS date using regex
    match = re.search(r'वि\.सं\. ([\u0900-\u097F]+) (\d{1,2}), (\d{4})', text)
    if not match:
        raise ValueError("Date not found or doesn't match the pattern.")

    month_nep = match.group(1)
    day = int(match.group(2))
    year = int(match.group(3))

    # Convert Nepali month to number
    month = nepali_months.get(month_nep)
    if not month:
        raise ValueError(f"Nepali month '{month_nep}' not recognized.")

    # Convert to NepaliDate object and then to Gregorian date
    bs_date = nepali_datetime.date(year, month, day)
    ad_date = bs_date.to_datetime_date()
    return ad_date

# Example use
text = 'दैनिक मूल्य बारे जानकारी\n- वि.सं. जेठ १८, २०८२'
print(nepali_date_to_ad(text))  # Output: 2025-06-01
```

```python
import re

# Mapping Devanagari digits to English
dev_to_eng = str.maketrans("०१२३४५६७८९", "0123456789")

def convert_nepali_currency_to_float(text):
    # Remove 'रू' and unnecessary whitespace
    cleaned = re.sub(r'रू|\s+', '', text)
    
    # Convert Devanagari to English digits
    english_number = cleaned.translate(dev_to_eng)

    # Convert to float
    try:
        return float(english_number)
    except ValueError:
        raise ValueError(f"Could not convert '{text}' to float.")

# Example
print(convert_nepali_currency_to_float("रू ४५.००"))   # Output: 45.0
```

```python
df['minimum'] = df['minimum'].apply(convert_nepali_currency_to_float)
df['maximum'] = df['maximum'].apply(convert_nepali_currency_to_float)
df['average'] = df['average'].apply(convert_nepali_currency_to_float)
```

```python
def convert_nepali_unit_to_english(text):
    unit_map = {
        'के.जी.': 'kg',
        'के जी': 'kg',
        'केजी': 'kg',
        'दर्जन': 'dozen',
        'प्रति गोटा': 'per piece',
        'गोटा': 'piece',
        'पोका': 'bundle',
        'किलो': 'kg',
        'लिटर': 'litre',
        'मुठा': 'bunch',
    }

    return unit_map.get(text.strip(), 'unknown')

# Example use
units = ['के.जी.', 'के जी', 'केजी', 'दर्जन', 'प्रति गोटा']
converted = [convert_nepali_unit_to_english(unit) for unit in units]
print(converted)
```
