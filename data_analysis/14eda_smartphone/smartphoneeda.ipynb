```python

x_df = df.select_dtypes(include=['int64','float64']).drop(columns='price')
```

```python
imputer = KNNImputer(n_neighbors=5)
```

```python
x_df_values = imputer.fit_transform(x_df)
```

```python
x_df = pd.DataFrame(x_df_values , columns=x_df.columns)
```
