```python
#Decorators
#python as 1st class citizen

def modify(func,num):
    return func(num)
def sq(num):
    return num**2

modify(sq,2)
```

```python
#Decorator Example
def my_decorator(func):
    def wrapper():
        print("*"*20)
        func()
        print("*"*20)
    return wrapper

def hello():
    print("Hello")

a = my_decorator(hello)
a()

```

```python
def my_decorator(func):
    def wrapper(*args):
        print("*" * 20)
        func(*args)
        print("*" * 20)
    return wrapper

@my_decorator
def sum_numbers(*num):
    temp = 0
    for i in num:
        temp += i
    print(temp)

# Now call it like this
sum_numbers(1, 2, 3, 5, 5)
```

```python
#Shortcut in decorators
#Decorator Example
def my_decorator(func):
    def wrapper():
        print("*"*20)
        func()
        print("*"*20)
    return wrapper

@my_decorator
def hello():
    print("Hello")

@my_decorator
def sum():
    print(124)

hello()
sum()
```

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print("The time taken by", func.__name__ , "is" , end-start,"secs")
    return wrapper

@timer
def hello():
    print("Hello")
    time.sleep(2)

@timer
def loop():
    for i in range(10000):
        with open("count.txt","a") as f:
            f.write(f" {i} ")


loop()
hello()
```
