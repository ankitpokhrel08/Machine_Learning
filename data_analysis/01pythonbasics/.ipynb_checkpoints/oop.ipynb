```python
#ATM class with oop
class Atm:
    def __init__(self):
        self.pin = ""
        self.amount = 0

        self.menu()

    def menu(self):
        user_input= input("""What do you like to do?
        1. Create new pin
        2. Deposit Amount
        3. Check Balance
        4. Withdraw
        5. Exit""")

        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.check_balance()
        elif user_input == "4":
            self.withdraw()
        else:
            print("Exited")

    def create_pin(self):
        if (self.pin == ""):
            self.pin = input("Enter your new Pin: ")
            print("Pin created successfully.")
        else:
            temp = input("Enter your old Pin: ")
            if temp == self.pin:
                self.pin = input("Enter your new Pin: ")
                print("Pin created successfully.")
            else:
                print("Pin didnot match.")

    def deposit(self):
        temp = input("Enter your Pin: ")
        if temp == self.pin:
            money = int(input("Enter how much you want to deposit: "))
            self.amount += money
        else:
            print("Pin didn't match")
    def check_balance(self):
        temp = input("Enter your Pin: ")
        if temp == self.pin:
            print("You have", self.amount)
        else:
            print("Pin didn't match")
    def withdraw(self):
        temp = input("Enter your Pin: ")
        if temp == self.pin:
            money = int(input("Enter how much you want to withdraw: "))
            self.amount -= money
        else:
            print("Pin didn't match")

    
```

```python
#ATM class with oop
class Atm2:
    def __init__(self):
        self.__pin = ""
        self.__amount = 0

        self.menu()

    def menu(self):
        user_input= input("""What do you like to do?
        1. Create new __pin
        2. Deposit __Amount
        3. Check Balance
        4. Withdraw
        5. Exit""")

        if user_input == "1":
            self.create___pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.check_balance()
        elif user_input == "4":
            self.withdraw()
        else:
            print("Exited")

    def create___pin(self):
        if (self.__pin == ""):
            self.__pin = input("Enter your new __Pin: ")
            print("__Pin created successfully.")
        else:
            temp = input("Enter your old __Pin: ")
            if temp == self.__pin:
                self.__pin = input("Enter your new __Pin: ")
                print("__Pin created successfully.")
            else:
                print("__Pin didnot match.")

    def deposit(self):
        temp = input("Enter your __Pin: ")
        if temp == self.__pin:
            money = int(input("Enter how much you want to deposit: "))
            self.__amount += money
        else:
            print("__Pin didn't match")
    def check_balance(self):
        temp = input("Enter your __Pin: ")
        if temp == self.__pin:
            print("You have", self.__amount)
        else:
            print("__Pin didn't match")
    def withdraw(self):
        temp = input("Enter your __Pin: ")
        if temp == self.__pin:
            money = int(input("Enter how much you want to withdraw: "))
            self.__amount -= money
        else:
            print("__Pin didn't match")

    
```

```python

#pass object into functions
class Customer:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

cust = Customer("Ankit","Male")
cust2 = Customer("Radha", "Female")

def greet(agent):
    if agent.gender == "Male":
        print("Hello", agent.name , "Sir")
    else:
        print("Hello", agent.name,"Ma'am")

greet(cust)
greet(cust2)
```

```python
#Inheritance
class User:
    def login(self):
        print("Login")

    def register(self):
        print("Register")

class Student(User):
    def enroll(self):
        print("Enrolled")

stu1 = Student()
```

```python
stu1.login()
stu1.enroll()
```
