```python
#Creating big files
import random
char1 = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
big_text = [char1[:random.randint(0,51)] for i in range(1000)]
with open("bigfile.txt",'w') as f:
    f.writelines(big_text)
```

```python
#loading big file with chunks
with open("bigfile.txt", 'r') as f:
    chunk = 10
    while True:
        s = f.read(chunk)
        if s:
            print(s,end="@gmail.com\n")
        else: break
```

```python
#Serialization and Deserialization
#List <---> JSON
import json
L= [1,2,3,4,"Ankit"]

with open("demo.json","w") as f:
    json.dump(L,f,indent = 4)
```

```python
#Deserilization
with open("demo.json","r") as fb:
    print(json.load(fb))
    fb.seek(0)
    print(type(json.load(fb)))
```

```python
with open("demo.json","r") as fb:
    a = json.load(fb)
    print(a)
    print(type(a)) 
```

```python
#This won't work because this is not the corect format of json
#Lis,Dictionary <---> JSON
import json
L= [1,2,3,4,"Ankit"]
D= {"Name":"Ankit" , "Age":16}
with open("demo.json","w") as f:
    json.dump(L,f,indent = 4)
    json.dump(D,f,indent = 4)
```

```python
#This is the correct way to do things!!!!
import json
L = [1, 2, 3, 4, "Ankit"]
D = {"Name": "Ankit", "Age": 16}

combined = {"list": L, "dict": D}

with open("demo.json", "w") as f:
    json.dump(combined, f, indent=4)
```

```python
#Passing objects into files 

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

myself = Person("Ankit",21)

def show_object(person):
    if isinstance(person,Person):
        return {"Name" : person.name,"Age":person.age}

with open("person.json","w") as f:
    json.dump(myself,f,default = show_object)

with open("person.json","r") as f:
    a = json.load(f)
    print(a)
    print(type(a))
```

```python
#Pickling and unpickling
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def show_object(self):
        return {"Name" : self.name,"Age": self.age}

myself = Person("Ankit",21)
```

```python
#pickle dump
import pickle
with open("person.pkl","wb") as f:
    pickle.dump(myself,f)
```

```python
#pickle load
import pickle 
with open("person.pkl","rb") as f:
    newpickle = pickle.load(f)
```
