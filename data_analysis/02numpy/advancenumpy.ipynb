## Comparing Numpy vs List

```python
#With python list
import time
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]
c = []
start = time.time()
for i in range(len(a)):
    c.append(a[i]+b[i])
end = time.time()
ltime = end - start
print(end-start)
```

```python
#with numpy array
import numpy as np
```

```python
d = np.arange(10000000)
e = np.arange(10000000,20000000)

start = time.time()
f = d+e
end = time.time()
ntime = end - start
```

```python
(ltime - ntime)
```

```python
#memory 
a = [i for i in range(10000000)]
b = np.arange(10000000,dtype = np.int32)
import sys
print(sys.getsizeof(a))
print(sys.getsizeof(b))
```

```python
#sigmoid using numpy
def sigmoid(array):
    return 1/(1+np.exp(-array))

a = np.random.randint(1,100,24).reshape(4,6)
sigmoid(a)
```

```python
#mean squared error
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
```

```python
def mse(actual,predicted):
    return np.mean((actual-predicted)**2)
```

```python
mse(actual,predicted)
```
