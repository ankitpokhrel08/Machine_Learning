```python
import numpy as np
```

```python
a = np.random.randint(1000000,20000000,3000000)
```

```python
import time 
start = time.time()
np.sort(a)
end= time.time()
print(end-start)
```

```python
import time 
start = time.time()
np.sort(a,kind = 'quick')
end= time.time()
print(end-start)
```

```python
import time 
start = time.time()
np.sort(a,kind = 'heap')
end= time.time()
print(end-start)
```

```python
#Factorial with cumpro
a = int(input("Enter a number whose factorial you want to see: "))
b = np.arange(1,a+1)
if a == 0:
    print("1")
else:
    print(np.cumprod(b)[-1])
```
