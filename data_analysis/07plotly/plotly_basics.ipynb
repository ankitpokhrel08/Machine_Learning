```python
nepse = pd.read_csv("nepse.csv", on_bad_lines='skip')
```

```python
nepse.head()
```

```python
nepse = nepse[['Symbol','Open Price','Close Price','Total Traded Quantity','Total Traded Value']]
```

```python
nepse.head(10)
```

```python
nepse['summary'] = nepse['Close Price'] - nepse['Open Price']
```

```python
nepse.head()
```

```python
nepse.rename(columns = {"Symbol" :'symbol', "Open Price" : 'open','Close Price' : 'close','Total Traded Quantity': "quantity","Total Traded Value":"value"  } , inplace = True)
```

```python
nepse.info()
```

```python
nepse[nepse['summary'] > 0]
```

```python

nepse['status'] = nepse['summary'].apply(lambda x: 
                                   'Profit' if x > 0 else 
                                   'Loss' if x < 0 else 
                                   'Hold')
```

```python
nepse[nepse['symbol'] == 'OMPL']
```

```python
px.treemap(nepse,path=[px.Constant("Today's Trade"),'status','symbol'] ,values='value',color='summary',color_continuous_scale=[(0, "red"), (1, "green")],color_continuous_midpoint=0)
```
